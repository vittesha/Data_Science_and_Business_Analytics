###Task 6: Prediction using Decision Tree Algorithm

```python
import numpy as np
import pandas as pd
import sklearn.datasets as datasets

df=pd.read_csv("Iris.csv")
df.head()
```

```python
var=['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']
x= df[var]
y= df.Species
```

```python
from sklearn import metrics
from sklearn.model_selection import train_test_split
train_data, test_data, train_label, test_label= train_test_split(x, y, test_size=0.2, random_state=0)
```

```python
train_data.shape, test_data.shape, train_label.shape, test_label.shape
```

Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
dtc= DecisionTreeClassifier()
dtc.fit(x,y)
```

Prediction on test data obtained using train_test_split

```python
pred_y=dtc.predict(test_data)
```

```python
comp=pd.DataFrame({'Actual':test_label, 'Predicted':pred_y})
print(comp)
```
